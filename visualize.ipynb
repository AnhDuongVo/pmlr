```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
i = 1
metric = ["accuracy", "balanced_accuracy", "f1_score", "average_precision_score"][i]

print(f"The metric is {metric}")
```

# Encrypted results vs original results

```python
clean_df = pd.read_csv("data/results_fhe_False.csv")
encrypted_df = pd.read_csv("data/results_fhe_True.csv")
```

```python

for test_split in ['A', 'B', 'C', 'D']:        
    plot_list = []
    for df in [encrypted_df, clean_df]:
        temp_df = df[clean_df['test_split'] == test_split]

        y_mean = 100 * temp_df[f"{metric}_mean"].to_numpy().astype(float)
        y_std = 100 * temp_df[f"{metric}_std"].to_numpy().astype(float)
        x_train_splits = temp_df["train_split"].values

        # Create a DataFrame for plotting
        plot_df = pd.DataFrame({
            'Train split': x_train_splits,
            metric: y_mean,
            'std': y_std
        })
        plot_list.append(plot_df)


    # Plot using Seaborn
    plt.figure(figsize=(10, 4))

    for i, plot_df in enumerate(plot_list):
        c = ['blue', 'red'][i]
        label=['Encrypted', 'Clean'][i]

        sns.lineplot(x='Train split', y=metric, data=plot_df, marker='o', c=c, linestyle='--', label = label)

        # Adding error bars
        plt.errorbar(x=plot_df['Train split'], y=plot_df[metric], yerr=plot_df['std'], fmt='o', c=c, capsize=5)

    plt.xlabel('Train split')
    plt.ylabel(metric)
    plt.title(f'{metric} for test split {test_split}')
    plt.grid(True)
    plt.xticks()  # Rotate x-axis labels if necessary
    plt.tight_layout()  # Adjust the layout to fit everything

    plt.show()
```

# For each split (original data)

```python
plot_list = []
df = clean_df

for test_split in ['A', 'B', 'C', 'D']:        
    temp_df = df[df['test_split'] == test_split]

    y_mean = 100 * temp_df[f"{metric}_mean"].to_numpy().astype(float)
    y_std = 100 * temp_df[f"{metric}_std"].to_numpy().astype(float)
    x_train_splits = temp_df["train_split"].values

    # Create a DataFrame for plotting

    plot_df = pd.DataFrame({
        'Train split': x_train_splits,
        metric: y_mean,
        'std': y_std
    })
    plot_list.append(plot_df)


# Plot using Seaborn
plt.figure(figsize=(14, 4))

for i, plot_df in enumerate(plot_list):
    c = ['blue', 'red', 'green', 'cyan'][i]
    label=['test split A', 'test split B', 'test split C', 'test split D'][i]

    sns.lineplot(x='Train split', y=metric, data=plot_df, marker='o', c=c, linestyle='--', label = label)

    # Adding error bars
    plt.errorbar(x=plot_df['Train split'], y=plot_df[metric], yerr=plot_df['std'], fmt='o', c=c, capsize=5)

plt.xlabel('Train split')
plt.ylabel(f'{metric} (%)')
plt.title(f'{metric} for all splits (original data)')
plt.grid(True)
plt.xticks()  
plt.tight_layout() 

plt.show()
```
